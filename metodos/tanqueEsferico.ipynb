```python
import numpy as np
import pandas as pd
import math
```

```python
def funcionIntegral(t, w):
    r = (math.pow(math.e, t) * math.sin(2*t + 8)) / (1 - t + t**2) 
    return r
```

```python
def rk1(h, t, w):
    r = h * funcionIntegral(t, w)
    return r
```

```python
def rk2(h, t, w, vk1):
    r = h * funcionIntegral(t + h/2, w + vk1/2)
    return r
```

```python
def rk3(h, t, w, vk2):
    r = h * funcionIntegral(t + h/2, w + vk2/2)
    return r
```

```python
def rk4(h, t, w, vk3):
    r = h * funcionIntegral(t + h, w + vk3)
    return r
```

```python

intervalo = np.linspace(2.1, 8, 60)

#print(intervalo)

h = 0.1
wi = 0

tablita = pd.DataFrame(columns=['ti', 'wi', 'k1', 'k2', 'k3', 'k4'])

vrk1 = rk1(h, 2, 0)
vrk2 = rk2(h, 2, 0, vrk1)
vrk3 = rk3(h, 2, 0, vrk2)
vrk4 = rk4(h, 2, 0, vrk3)

tablita.loc[len(tablita)] = (2, 0, vrk1, vrk2, vrk3, vrk4)

for i in intervalo:
    wi = wi + (1/6)*(vrk1 + 2*vrk2 + 2*vrk3 + vrk4)
    vrk1 = rk1(h, i, 0)
    vrk2 = rk2(h, i, 0, vrk1)
    vrk3 = rk3(h, i, 0, vrk2)
    vrk4 = rk4(h, i, 0, vrk3)
    tablita.loc[len(tablita)] = (i, wi, vrk1, vrk2, vrk3, vrk4)

tablita
```
