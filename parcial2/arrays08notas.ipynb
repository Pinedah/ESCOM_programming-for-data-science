# Operaciones con arreglos en Numpy

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
grados = np.array([i for i in range(0, 361, 1)], dtype='float') 
grados
```

```python
# Multiplicar un arreglo completo de flotantes, por una constante
# (Como multiplicar un escalar por un vector)
# Operacion bitwise (Elemento por elemento)
np_radianes = np.array(grados * (math.pi / 180), dtype='float')
np_radianes
```

```python
fsin = np.sin(np_radianes)
fsin
```

```python
radianes2 = np.radians(np.array([i for i in range(0, 191, 10)], dtype='float').reshape(5,4))
radianes2
```

```python
# graficar la funcion seno
plt.plot(grados, fsin, linestyle='-', linewidth='2')
```

```python
fcos = np.cos(radianes2)
fcos
```
